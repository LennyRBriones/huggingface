```python
text = """
Los presentes términos y condiciones de Aeroenlaces Nacionales, S.A. de C.V. ("Viva Aerobus") tienen como objetivo dar a conocer a nuestros Pasajeros las condiciones generales de contratación bajo las cuales se presta el servicio de transportación aérea (el “Servicio”) (los “Términos y Condiciones”).

Con el fin de mejorar el Servicio proporcionado a nuestros Pasajeros, Viva Aerobus podrá cambiar y/o modificar en cualquier tiempo los presentes Términos y Condiciones ya sea parcial o totalmente, siendo aplicables los Términos y Condiciones que se encuentren publicados en el sitio web www.vivaaerobus.com (el “Sitio en Internet”) al momento de la prestación del Servicio. No obstante, lo anterior, en aquellos casos en los que la modificación a los Términos y Condiciones pudiera causar una afectación al Pasajero relacionada con la prestación del Servicio, Viva Aerobus le notificará a dicho Pasajero la modificación en los Términos y Condiciones a través del Sitio en Internet www.vivaaerobus.com, así como a través de uno de los medios de contacto que el Pasajero hubiera designado al momento de realizar su Reservación.

En aquellos casos en los que el Pasajero considere que la modificación pueda afectar su Reservación (como dicho término se define más adelante) deberá ponerse en contacto con el Call Center (como dicho término se define más adelante) de Viva Aerobus y en caso de ser aplicable, se podrán aplicar las compensaciones que correspondan según las Políticas de Compensación (término que se define más adelante) de Viva Aerobus.

Sin importar que existan modificaciones a los presentes Términos y Condiciones, Viva Aerobus respetará las condiciones vigentes al momento en el que se haya realizado el pago de la Reservación contratada por el Pasajero, salvo los casos y con las excepciones establecidas en la legislación aplicable.

Los títulos alusivos que aparecen en los presentes Términos y Condiciones son utilizados únicamente como referencia, sin que de ellos se desprendan derechos u obligaciones.

"""

summarization_es(text, min_length=30, max_length=150)
```

```python
text_classification = pipeline(task="text-classification", model="pysentimiento/robertuito-sentiment-analysis")
```

```python
texts = ["Tengo muchas ganas de ir al baño pero está muy lejos y está ocupado",
         "Me parece fascinante cómo puede darse a entender este material"]

text_classification(texts)
```
