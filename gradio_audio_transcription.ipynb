```python
from transformers import pipeline
import gradio as gr

model = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-large-xlsr-53-spanish")

def transcribe(audio):
  text = model(audio)["text"]
  return text
```

```python
gr. Interface(
    fn = transcribe,
    inputs=[gr.Audio(source="microphone", type="filepath")],
    outputs=["textbox"]
).launch()
```

```python
title = "Test model speech transcption"
description= "Loading pretrained model"

gr.Interface.load(
    "facebook/wav2vec2-large-xlsr-53-spanish",
    #the pretrained model to use
    inputs = gr.Audio(label="Try to speech here (graba un audio en este apartado)"),
    title = title,
    description = description
).launch()
```
